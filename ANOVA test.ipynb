```python
import seaborn as sns
import pandas as pd
from scipy.stats import f_oneway
```

# Load the dataset

```python
df = sns.load_dataset('titanic')
df
```

```python
df['pclass'].unique()
```

```python
df = df[['age','pclass']].dropna()
df
```

```python
class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']
```

```python
f_stats , p_value = f_oneway(class_1 , class_2 , class_3)
f_stats
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("reject the null hypothesis and there is a significant difference in atleast 1 passenger class")
else:
    print("there is no  significant difference")
```
